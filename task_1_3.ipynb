# Task 1.3 — What the Paper Claims to Improve
**Paper:** Incremental Learning of Temporally-Coherent Gaussian Mixture Models


## Main Baseline

The primary baseline the paper compares against is **MDL-EM (Expectation-Maximisation with the Minimal Description Length criterion)** — specifically the standard offline/batch EM algorithm applied with MDL model order selection as described in Rissanen [12] and used widely in computer vision GMM fitting.

---

## Limitation of the Baseline

The batch MDL-EM approach requires all data to be stored in memory simultaneously before fitting begins. This is impractical for real-time applications (e.g., face modelling from live video) where data arrives continuously, one point at a time, and storing the full history either exhausts memory or introduces unacceptable latency. Additionally, standard EM cannot dynamically update model complexity as new structure emerges; it must be rerun from scratch on all data whenever the dataset grows.

---

## How the Proposed Method Overcomes It

The TC-GMM method overcomes this limitation by (a) keeping only two GMMs in memory — the current estimate and the Historical GMM — with no raw data storage, (b) updating parameters in closed form per new point (Eq. 7–8) in O(M·D²) time, and (c) dynamically adjusting the number of components using the Historical GMM snapshot and MDL-based expected description length, so model complexity can grow or shrink online without re-processing old data.

---

## Condition Under Which the Paper's Method Would Not Outperform the Baseline

The TC-GMM method would likely not outperform batch MDL-EM when the data arrives in a completely random, non-temporally-coherent order — for example, when all observations are shuffled i.i.d. samples with no smoothness in time. In this case, the temporal coherence assumption (Section 2.1) is violated, the approximation in Eq. (6) becomes inaccurate, the Historical GMM accumulates errors rather than meaningful drift signals, and the component split/merge decisions become unreliable. The batch method, having access to all data, would produce a more accurate final model under these conditions.
