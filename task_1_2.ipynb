# Task 1.2 — Key Assumptions
**Paper:** Incremental Learning of Temporally-Coherent Gaussian Mixture Models


## Assumption 1

**Assumption:** Consecutive data points are temporally coherent — the distance between x_t and x_{t+1} follows a unimodal distribution p_S peaked at a small but non-zero value.

**Why the method needs it:** The fixed-complexity update (Eq. 7–8) approximates p*(i|x_j) ≈ p(i|x_j) (Eq. 6) — i.e., that adding one new point barely changes the responsibilities of existing data. This approximation only holds when successive points are similar enough that one new point does not dramatically shift the model. If successive points are uncorrelated and could lie anywhere in space, Eq. 6 breaks down and the updates become inaccurate.

**Violation scenario:** A sensor stream that alternates randomly between two very distant locations, or a video system that drops frames such that consecutive faces are completely different identities — each new point can lie in a completely different region from the previous, making the TC assumption fail.

**Paper reference:** Section 2.1, TC-GMM definition and Figure 1(a); Section 2.3, paragraph justifying Eq. (6).

---

## Assumption 2

**Assumption:** Between the Historical GMM and the current GMM, no permanent model order change has occurred — meaning there is a guaranteed 1-to-1 correspondence between components of both GMMs.

**Why the method needs it:** Equations (9) and (10) compute the "difference component" by subtracting historical component parameters from current ones. This subtraction only makes sense if component i in the historical GMM corresponds exactly to component i in the current GMM. If any component was created or destroyed between the two snapshots, the correspondence breaks and the difference component is meaningless or misleading.

**Violation scenario:** A dataset where two separate cluster-formation events happen in quick succession before the algorithm has a chance to lock in the historical GMM. The method could confuse two unrelated component drifts as a single new cluster signal.

**Paper reference:** Section 2.4, second paragraph: "as, by definition, no permanent model order changes occurred between the Historical and current GMMs, they have the same number of components and the 1-1 correspondence between them is known."

## Assumption 3

**Assumption:** The true underlying data distribution is well-approximated by a Gaussian Mixture Model with a finite and relatively small number of components.

**Why the method needs it:** MDL-based merging (Eq. 11–12) penalises complexity using the count of free parameters N_E, which grows quadratically with dimensionality D and linearly with component count M. If the true distribution requires many components or is not Gaussian (e.g., a multimodal distribution with heavy tails, ring-shaped clusters), the MDL criterion will over-penalise the split and collapse components incorrectly.

**Violation scenario:** A 2D dataset of points sampled uniformly from a thin ring (annulus) — no finite GMM can accurately capture this; the method will keep merging all components into one large Gaussian and achieve poor density estimates regardless of how much data arrives.

**Paper reference:** Section 1, paragraph 1 ("The GMM is a semi-parametric method for high-dimensional density estimation"); Section 2, Eq. (1); MDL derivation Section 2.5.
